## ONE HOT ENCODING

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame({'color':['red','green','blue','green','red','blue','red']})
```

```python
df
```

```python
## creating instance of OHE
ohe=OneHotEncoder()
```

```python
encoded=ohe.fit_transform(df[['color']]).toarray()
```

```python
encoded
```

```python
df2=pd.DataFrame(encoded,columns=ohe.get_feature_names_out())
df2
```

```python
pd.concat([df,df2],axis=1)
```
