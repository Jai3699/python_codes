```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.linspace(0,10,200)
```

```python
x
```

```python
y=np.sin(x)
```

```python
y
```

```python
plt.plot(x,y)
plt.ylabel('y')
plt.xlabel('x')
plt.title('sin graph')
```

```python
a=np.random.rand(50)
b=np.random.rand(50)
```

```python
plt.scatter(a,b)
```

```python
plt.scatter(a,b,color='b')
```

```python
plt.scatter(a,b,color='b',s=700)
```

```python
plt.figure(figsize=(10,8))
colors=np.random.rand(50)
sizes=2000*np.random.rand(50)
plt.scatter(a,b,c=colors,s=sizes,alpha=.8)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Random Data')
```

```python
x=['a','b','c','d','e']
```

```python
y=np.random.rand(5)
```

```python
plt.bar(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Bar-plot')
```

```python
plt.barh(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Bar-plot')
```

```python
plt.figure(figsize=(8,5))
plt.plot(x,y,'--r')
plt.grid()
plt.ylabel('y')
plt.xlabel('x')
plt.title('sin graph')
```

```python
plt.figure(figsize=(10,8))
colors=np.random.rand(50)
sizes=2000*np.random.rand(50)
plt.scatter(a,b,c=colors,s=sizes,alpha=.8,marker='v')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Random Data')
```
