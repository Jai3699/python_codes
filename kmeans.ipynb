```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs ## blobs is a dataset for clusterng
```

```python
x,y=make_blobs(n_samples=1000,centers=3,n_features=2,random_state=23) ## we are taking three centers that means three clusters
```

```python
## in cluster we dont need y but make_blobs gives y that's why we are taking it here further we will ignore it
```

```python
x.shape
```

```python
plt.scatter(x[:,0],x[:,1],c=y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
from sklearn.cluster import KMeans
```

```python
## manually for finding the k value using elbow method
```

```python
wcss=[]
for k in range(1,11):
    kmeans=KMeans(n_clusters=k,init='k-means++')
    kmeans.fit(x_train)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss)
plt.xticks(range(1,11))
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
```

```python
kmeans=KMeans(n_clusters=3,init='k-means++')
```

```python
kmeans.fit(x_train) 
```

```python
kmeans.inertia_
```

```python
kmeans.labels_ ## can also find using kmeans.predict
```

```python
y_labels=kmeans.predict(x_train)
```

```python
y_labels
```

```python
plt.scatter(x_train[:,0],x_train[:,1],c=y_labels) ## this means our assumption for k value is right bcz we are getting the k value 
## equal to three whcih is good
```

```python
y_test_labels=kmeans.predict(x_test) ## for test we only predict
```

```python
plt.scatter(x_test[:,0],x_test[:,1],c=y_test_labels)
```

```python
## performance matrix
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(x_train,y_labels)
```

```python
## finding score for all k values
```

```python
silhoutte_score=[]
for k in range(2,11):
    kmeans=KMeans(n_clusters=k,init='k-means++')
    kmeans.fit(x_train)
    score=silhouette_score(x_train,kmeans.labels_)
    silhoutte_score.append(score)
```

```python
silhoutte_score
```

```python
plt.plot(range(2,11),silhoutte_score)
plt.xlabel('Number of Cluster')
plt.ylabel('Accuracy')
```
