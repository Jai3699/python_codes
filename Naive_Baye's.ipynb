## Naive Baye's 

```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_iris
```

```python
x,y=load_iris(return_X_y=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb=GaussianNB()
```

```python
gnb.fit(x_train,y_train)
```

```python
y_pred=gnb.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
import seaborn as sns
```

```python
tip=sns.load_dataset('tips')
```

```python
tip.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe=OneHotEncoder()
```

```python
encoded=ohe.fit_transform(tip[['sex','smoker','day']])
```

```python
encoded_data=encoded.toarray()
```

```python
df_encoded=pd.DataFrame(encoded_data,columns=ohe.get_feature_names_out())
```

```python
df_final=pd.concat([tip,df_encoded],axis=1)
```

```python
df_final
```

```python
df_final=df_final.drop(columns=['sex','smoker','day'])
```

```python
df_final.head()
```

## we have to find time based on other features

```python
x=df_final.drop(['time'],axis=1)
```

```python
x
```

```python
y=df_final['time']
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb=GaussianNB()
```

```python
gnb.fit(x_train,y_train)
```

```python
y_pred=gnb.predict(x_test)
```

```python
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
